```python
#top 10 most common genres
genres= df['listed_in'].dropna().str.split(",")
```

```python
genres
```

```python
flat_genres=[]
for sublist in genres:
    for genre in sublist:
        flat_genres.append(genre)
   
```

```python
from collections import Counter
```

```python
genre_count= Counter(flat_genres)
top_10_genres=genre_count.most_common(10)
```

```python
genre_df=pd.DataFrame(top_10_genres,columns=['Genre','Count'])
genre_df
```

```python
df['duration_mins']=df['duration'].str.extract("(\d+)").astype(float)
df['Seasons']=df.loc[df['type']=='TV Show','duration'].str.extract("(\d+)").astype(float)
avg_movie_dur=df[df['type']=='Movie']['duration_mins'].mean()
avg_tv_shows=df[df['type']=='TV Show']['Seasons'].mean()
print(avg_movie_dur)
print(avg_tv_shows)
```

```python
avg_duration=df.groupby('type')['duration_mins'].mean()
avg_duration.plot(kind='bar',color=['orange','red'])
plt.title("Average duration by Type")
plt.xlabel('duration')
plt.ylabel('movies in duration and shows in seasons')
plt.show()
```
